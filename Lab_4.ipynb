```python
t = np.arange(0,2.5,0.1)
y1 = np.sin(np.pi*t)
y2 = np.sin(np.pi*t+np.pi/2)
y3 = np.sin(np.pi*t-np.pi/2)
plt.plot(t,y1,'r*',t,y2,'g^',t,y3,'ys')
```

```python
df = pd.DataFrame(dict(
    x = np.arange(0,2.5,0.1),
    y = np.sin(np.pi*t)
))
fig = px.line(df, x="x", y="y", title="Sin Wave") 
fig.show()
```

```python
t = np.arange(0,2.5,0.1)
y1 = np.sin(np.pi*t)
y2 = np.sin(np.pi*t+np.pi/2)
y3 = np.sin(np.pi*t-np.pi/2)
plt.subplot(311)
plt.plot(t,y1)
plt.subplot(312)
plt.plot(t,y2)
plt.subplot(313)
plt.plot(t,y3)
```

```python
dx = 0.01
dy = 0.01
x = np.arange(-2.0 , 2.0 , dx)
y = np.arange(-2.0,2.0, dy)
X,Y = np.meshgrid(x,y)
def f(x,y):
  return (1 - y**5 + x**5) * np.exp(-x**2-y**2)

C = plt.contour(X,Y , f(X,Y) , 8 , colors = "black")
plt.contourf(X,Y , f(X,Y),8)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
dx = 0.01
dy = 0.01
x = np.arange(-2.0 , 2.0 , dx)
y = np.arange(-2.0,2.0, dy)
X,Y = np.meshgrid(x,y)
def f(x,y):
  return (1 - y**5 + x**5) * np.exp(-x**2-y**2)

C = plt.contour(X,Y , f(X,Y) , 8 , colors = "black")
plt.contourf(X,Y , f(X,Y),8)
```
